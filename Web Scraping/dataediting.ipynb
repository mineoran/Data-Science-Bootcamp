```python
import pandas as pd
df = pd.read_csv('data.csv')
```

```python
df.head(5)
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
print (df.shape)
```

```python
df = df.join(df['location'].str.split(',', 1, expand=True).rename(columns={0:'City', 1:'State'}))
```

```python
df.head()
```

```python
def manager(x):
    if 'Manager' in x:
        return 1
    return 0

df['Manager'] = df['title'].apply(manager)
```

```python
df[df.Manager != 0]
```

```python
def engineer(x):
    if 'Engineer' in x:
        return 1
    return 0

df['Engineer'] = df['title'].apply(engineer)
```

```python
df[df.Engineer != 0]
```

```python
df['job salary'] = df['job salary'].str.replace(',', '')
df.head()
```

```python
df.to_csv('dataanalyst.csv', encoding='utf-8')
```
